```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
iris = load_iris()
x,y = iris.data,iris.target
```

```python
# === Model definition ===

class LogisticRegression:
  def __init__(self,learning_rate=0.001,n_inters=1000):
    self.learning_rate = learning_rate
    self.n_inters = n_inters
    self.weights = None
    self.bias = None

  def sigmoid(self,X):
    return 1 / (1 + np.exp(-X))

  def fit(self,X,y):
    n_samples , n_features = X.shape
    self.weights = np.zeros(n_features)
    self.bias = 0

    for _ in range(self.n_inters):
      linear_equation = np.dot(X,self.weights) + self.bias # y = ax + b (LOSS)
      y_hat = self.sigmoid(linear_equation)
      weights_derivate = (1/n_samples) * np.dot(X.T,(y_hat - y))
      bias_derivate = (1/n_samples) * np.sum(y_hat - y)

      self.weights -= self.learning_rate * weights_derivate
      self.bias -= self.learning_rate * bias_derivate

  def predict(self,X):
    linear_equation = np.dot(X,self.weights) + self.bias # y = ax + b (LOSS)
    y_hat = self.sigmoid(linear_equation)
    predicted = np.where(y_hat > 0.5,1,0) # if else condition
    return np.array(predicted)
```

```python
# === Metrics ===

def accuracy(y_true,y_hat):
  return np.sum(y_hat == y_true) / len(y_true)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,stratify=y,random_state=42,train_size=0.8)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# === Baseline ===

baseline = np.mean(y_train)
y_baseline = np.full_like(y_test,baseline)

print(f"Baseline accuracy : {accuracy(y_test,y_baseline):.2f}")
```

```python
model = LogisticRegression(learning_rate=0.001,n_inters=1000)
model.fit(x_train,y_train)
```

```python
y_hat = model.predict(x_test)

print(f"Logistic model accuracy : {accuracy(y_test,y_hat):.2f}")
```
